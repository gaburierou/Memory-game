```python
from tkinter import *
from tkinter import messagebox
import random

version = "0.3"

class Memory(Frame):
    def __init__(self, master):
        Frame.__init__(self, master)
        self.areas = None
        self.pinned = None
        self.joint = None
        self.count = None        
        self.widgets()        
    
    def start(self):#start variable
        self.areas = ['A!', 'A?', 'B!','B?',
                     'C!', 'C?','D!', 'D?']        
        self.pinned = []       
        self.joint = [self.areas[1],self.areas[3],self.areas[5],
                      self.areas[7]]       
        self.count = 3                
        self.shuffle = random.sample(self.areas, len(self.areas))
        self.one_show["text"] = "   "
        self.two_show["text"] = "   "
        self.three_show["text"] = "   "
        self.four_show["text"] = "   "
        self.five_show["text"] = "   "
        self.six_show["text"] = "   "
        self.seven_show["text"] = "   "
        self.eight_show["text"] = "   "

        self.show_buttons()
        self.memory2()
        
    def engine(self, x):#game 
        win = False
        if x == self.areas[self.get_index(self.areas, x)] and x not in self.pinned:            
            self.pinned.append(x)
            self.show_pinned(x)               
            if x in self.joint:
                if self.areas[self.get_index(self.areas, x)] and self.areas[self.get_index(self.areas, x) - 1] in self.pinned:                       
                    self.areas[self.get_index(self.areas, x) -1] = 0
                    self.areas[self.get_index(self.areas, x)] = 0
                    self.show_pinned(self.shuffle[self.get_index(self.shuffle, self.pinned[0])])
                    self.show_pinned(self.shuffle[self.get_index(self.shuffle, self.pinned[1])])
                    self.pinned.clear()                        
                else:
                    if len(self.pinned) > 1:
                        self.memory3()                        
                        self.hide_pinned(self.pinned[0])
                        self.hide_pinned(self.pinned[1])
                        self.pinned.clear()
                        self.count -= 1
                        self.show_count["text"] = f"score: {self.count}"                       
            else:
                if self.areas[self.get_index(self.areas, x)] and self.areas[self.get_index(self.areas, x) + 1] in self.pinned:                                               
                    self.areas[self.get_index(self.areas, x) + 1] = 0
                    self.areas[self.get_index(self.areas, x)] = 0
                    self.show_pinned(self.shuffle[self.get_index(self.shuffle, self.pinned[0])])
                    self.show_pinned(self.shuffle[self.get_index(self.shuffle, self.pinned[1])])
                    self.pinned.clear()                        
                else:
                    if len(self.pinned) > 1:
                        self.memory3()
                        self.hide_pinned(self.pinned[0])
                        self.hide_pinned(self.pinned[1])
                        self.pinned.clear()
                        self.count -= 1
                        self.show_count["text"] = f"score: {self.count}"                                               
            if any(isinstance(s, str) for s in self.areas):
                win = False
            else:
                messagebox.showinfo("Congrats","YOU WIN!")
                self.start()
                
            if self.count <= 0:
                messagebox.showinfo("Info","You lose :(")
                self.start()
 
        else:
            messagebox.showinfo("Info","Invalid value!")             

    def get_index(self, itens, n):     
        for i in itens:
            if n not in itens:
                return -1
            else:
                return itens.index(n)

    def show_pinned(self, x):#show selected
        if x == self.shuffle[0]:
            self.one_show["text"] = self.shuffle[0]
        if x == self.shuffle[1]:
            self.two_show["text"] = self.shuffle[1]
        if x == self.shuffle[2]:
            self.three_show["text"] = self.shuffle[2]
        if x == self.shuffle[3]:
            self.four_show["text"] = self.shuffle[3]
        if x == self.shuffle[4]:
            self.five_show["text"] = self.shuffle[4]
        if x == self.shuffle[5]:
            self.six_show["text"] = self.shuffle[5]
        if x == self.shuffle[6]:
            self.seven_show["text"] = self.shuffle[6]
        if x == self.shuffle[7]:
            self.eight_show["text"] = self.shuffle[7]
            
    def hide_pinned(self, x):#hide wrong
        if x == self.shuffle[0]:
            self.one_show["text"] = "   "
        if x == self.shuffle[1]:
            self.two_show["text"] = "   "
        if x == self.shuffle[2]:
            self.three_show["text"] = "   "
        if x == self.shuffle[3]:
            self.four_show["text"] = "   "
        if x == self.shuffle[4]:
            self.five_show["text"] = "   "
        if x == self.shuffle[5]:
            self.six_show["text"] = "   "
        if x == self.shuffle[6]:
            self.seven_show["text"] = "   "
        if x == self.shuffle[7]:
            self.eight_show["text"] = "   "      
            
    def widgets(self):
        #show game
        self.frame1 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame1.grid(row=0)
        
        self.frame2 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame2.grid(row=0, column=1)
        
        self.frame3 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame3.grid(row=0, column=2)        
        
        self.frame4 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame4.grid(row=0, column=3)
        
        
        
        self.frame5 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame5.grid(row=1, column=0)
        
        self.frame6 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame6.grid(row=1, column=1)       
        
        self.frame7 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame7.grid(row=1, column=2)
        
        self.frame8 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame8.grid(row=1, column=3)
        
        self.frame13 = Frame(self, relief = SUNKEN, borderwidth=10)
        self.frame13.grid(row=3, column=0)
        
        self.frame14 = Frame(self)
        self.frame14.grid(row=3, column=1)
        
               ######################                
      
        self.one_show = Label(self.frame1, font=("Verdana", 12), text="   ", borderwidth=10,padx=20,pady=10)
        self.one_show.grid(row=0, sticky="NWNESWSE")
        
        self.two_show = Label(self.frame2, font=("Verdana", 12),text="   ", borderwidth=10,padx=20,pady=10)
        self.two_show.grid(row=0, sticky="NWNESWSE")
        
        self.three_show = Label(self.frame3, font=("Verdana", 12), text="   ", borderwidth=10,padx=20,pady=10)
        self.three_show.grid(row=0, sticky="NWNESWSE")
              
        self.four_show = Label(self.frame4, font=("Verdana", 12), text="   ", borderwidth=10,padx=20,pady=10)
        self.four_show.grid(row=0, sticky="NWNESWSE")
        
        
        
        self.five_show = Label(self.frame5, font=("Verdana", 12), text="   ", borderwidth=10,padx=20,pady=10)
        self.five_show.grid(row=0, sticky="NWNESWSE")
        
        self.six_show = Label(self.frame6, font=("Verdana", 12), text="   ", borderwidth=10,padx=20,pady=10)
        self.six_show.grid(row=0, sticky="NWNESWSE")        
        
        self.seven_show = Label(self.frame7, font=("Verdana", 12), text="   ", borderwidth=10,padx=20,pady=10)
        self.seven_show.grid(row=0, sticky="NWNESWSE")
        
        self.eight_show = Label(self.frame8, font=("Verdana", 12), text="   ", borderwidth=10,padx=20,pady=10)
        self.eight_show.grid(row=0, sticky="NWNESWSE")
        
        #buttons
        
        self.start_button = Button(self.frame13,font=("System", 12), text="start", padx=10, pady=0, command=lambda: self.start())
        self.start_button.grid(row=0)
        
    def show_buttons(self):        
        self.button1 = Button(self.frame1, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.engine(self.shuffle[0]))
        self.button1.grid(row=1, column=0)
        
        self.button2 = Button(self.frame2, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.engine(self.shuffle[1]))
        self.button2.grid(row=1, column=0)
        
        self.button3 = Button(self.frame3, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.engine(self.shuffle[2]))
        self.button3.grid(row=1, column=0)
        
        self.button4 = Button(self.frame4, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.engine(self.shuffle[3]))
        self.button4.grid(row=1, column=0)
        
        self.button5 = Button(self.frame5, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.engine(self.shuffle[4]))
        self.button5.grid(row=1, column=0)
        
        self.button6 = Button(self.frame6, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.engine(self.shuffle[5]))
        self.button6.grid(row=1, column=0)
        
        self.button7 = Button(self.frame7, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.engine(self.shuffle[6]))
        self.button7.grid(row=1, column=0)
        
        self.button8 = Button(self.frame8, font=("System", 6), text="x", padx=10, pady=0, command=lambda: self.engine(self.shuffle[7]))
        self.button8.grid(row=1, column=0)
        
        #count ( not a button :3 )
        self.show_count = Label(self.frame14, font=("System", 12), text=f"score: {self.count}", padx=10, pady=0)
        self.show_count.grid(row=0, column=1)
        
    def memory2(self):#beginning game window
        memory2 = Tk()
        memory2.title("Show...")
        memory2.resizable(0,0)
        app = memory2.grid()
        
        f1 = Frame(memory2, relief = SUNKEN, borderwidth=10)
        f1.grid(row=0)
        f2 = Frame(memory2, relief = SUNKEN, borderwidth=10)
        f2.grid(row=0, column=1)
        f3 = Frame(memory2, relief = SUNKEN, borderwidth=10)
        f3.grid(row=0, column=2)
        f4 = Frame(memory2, relief = SUNKEN, borderwidth=10)
        f4.grid(row=0, column=3)
        
        f5 = Frame(memory2, relief = SUNKEN, borderwidth=10)
        f5.grid(row=1, column=0)
        f6 = Frame(memory2, relief = SUNKEN, borderwidth=10)
        f6.grid(row=1, column=1)
        f7 = Frame(memory2, relief = SUNKEN, borderwidth=10)
        f7.grid(row=1, column=2)
        f8 = Frame(memory2, relief = SUNKEN, borderwidth=10)
        f8.grid(row=1, column=3)
        
        L_1 = Label(f1, font=("Verdana", 12), text= self.shuffle[0], borderwidth=10,padx=20,pady=10)
        L_1.grid(row=0)
        L_2 = Label(f2, font=("Verdana", 12), text= self.shuffle[1], borderwidth=10,padx=20,pady=10)
        L_2.grid(row=0)
        L_3 = Label(f3, font=("Verdana", 12), text= self.shuffle[2], borderwidth=10,padx=20,pady=10)
        L_3.grid(row=0) 
        L_4 = Label(f4, font=("Verdana", 12), text= self.shuffle[3], borderwidth=10,padx=20,pady=10)
        L_4.grid(row=0) 
        L_5 = Label(f5, font=("Verdana", 12), text= self.shuffle[4], borderwidth=10,padx=20,pady=10)
        L_5.grid(row=0) 
        L_6 = Label(f6, font=("Verdana", 12), text= self.shuffle[5], borderwidth=10,padx=20,pady=10)
        L_6.grid(row=0) 
        L_7 = Label(f7, font=("Verdana", 12), text= self.shuffle[6], borderwidth=10,padx=20,pady=10)
        L_7.grid(row=0) 
        L_8 = Label(f8, font=("Verdana", 12), text= self.shuffle[7], borderwidth=10,padx=20,pady=10)
        L_8.grid(row=0)         
        memory2.after(3000, memory2.destroy)
        
    def memory3(self):#show wrong choice
        memory3 = Tk()
        memory3.title("Info")
        memory3.resizable(0,0)
        app = memory3.grid()
        
        F1 = Frame(memory3, relief = SUNKEN, borderwidth=10)
        F1.grid(row=0)
        F2 = Frame(memory3, relief = SUNKEN, borderwidth=10)
        F2.grid(row=0, column=1)
        F3 = Frame(memory3)
        F3.grid(row=1)
        
        l1 = Label(F1, font=("Verdana", 12), text= self.pinned[0], borderwidth=10,padx=20,pady=10)
        l1.grid(row=0, sticky="NWNESWSE")        
        l2 = Label(F2, font=("Verdana", 12),text= self.pinned[1], borderwidth=10,padx=20,pady=10)
        l2.grid(row=0, sticky="NWNESWSE")
        l3 = Label(F3, font=("Verdana", 9),text="WRONG!", borderwidth=10,padx=2,pady=1)
        l3.grid(row=0, sticky="NWNESWSE")                
        memory3.after(1500, memory3.destroy)
             
memory = Tk()
memory.title(f"Memory Game {version}")
memory.resizable(0,0)
app = Memory(memory).grid()
memory.mainloop()
```
